## 1. Introduction

In this Jupyter notebook, we provide an alternative to the popular modern portfolio theory (MPT) approach to optimizing asset allocations. In contrast to MPT, where financial risk is modelled by the volatility, i.e. standard errors, of the predicted returns, we choose to characterize this risk more explicitly by predicting an empirical joint distribution of returns and formulating the optimization problem with the objective to choose asset allocations to maximize the mean of this distribution and with the constraints that the choice of asset allocation ensures that certain values at risk (VaR) measured on this empirical distritbuion may not be violated. The main reason for this approach is to address one of the major drawbacks of MPT, namely not necessarily capturing the behvaior of the possibly heavy tails of the return distribution, thus underestimating the actual risk of an asset.

The overall approach can be summarized as follows:
1. We build a time series model based on historical asset returns, such that the residual errors of the model are independently and identically distributed (i.i.d.).
2. We use the model and the residual errors to generate bootstrapped predictions, i.e. predicting the next return value using the time series model and randomly sampling from the i.i.d. residual errors. In partciluar, making such predictions repetitively, we can generate a number of sample paths. The cumulative returns of each sample path can then be combined into an empirical distribution of predicted returns.
3. We formulate a mixed integer linear programming (MILP) problem that uses the empirical distribution of predicted returns to find the asset allocation that maximizes the mean of this distribution subject to one or several VaR constraints.

In the following sections, we first prepare some publicly available data and then go through these steps sequentially.

## 3. Data preparation
The function below reads in the data and removes rows with empty entries (corresponding to weekends and holidays on which the assets were not traded). The function returns the log-returns $\ln(p_{t+1}/p_t)$ for all time steps $t$, where $p_t$ is the closing price at time step $t$.

````python
def loadData(files):
    """
    Reads csv "files", fills in missing dates and close prices, computes daily log returns and outputs a pandas data frame.
    """
    dataFrames_in = []
    start_dates = [] 
    end_dates =[] #used to find greatest common trade period
    
    for file in files:
        df_in = pd.read_csv('C:/Users/worc3920/Documents/GitHub/EmpiricalPortfolio/data/'+file+'.csv', \
                            usecols=["Date","Close"]) #file path may depend on your project
        df_in = df_in.rename(columns={"Close": "Close_" + file}) #give unique column name according to file name
        df_in["Date"]=pd.to_datetime(df_in["Date"], infer_datetime_format=True)
        dataFrames_in.append(df_in)
        start_dates.append(df_in.loc[0,"Date"]) #this file's first entry is on this date
        end_dates.append(df_in.loc[df_in.index[-1],"Date"]) #and this is the last date
    
    #Compute greatest common trade period
    df_start = max(start_dates) #select latest common start date
    df_end = min(end_dates) #select earliest common end date, hence data available for all selected dates and assets
    df = pd.DataFrame()
    df["Date"] = pd.date_range(start=df_start, end=df_end)
    
    #Combine into single data frame
    for i in range(len(files)):
        df = df.merge(dataFrames_in[i], on='Date', how='inner') #combine data frames, "inner" to avoid nan values
    
    #Drop rows with missing data
    df = df.dropna()
    df = df.reset_index(drop=True)
    
    #Compute log return, assuming that negative final values will be replaced by 10**(-6)
    for file in files:
        for i in range(len(df)-1):
            df.loc[i, "Return_" + file] = np.log(np.max([10**(-6),df.loc[i+1, "Close_"+file]])/df.loc[i, "Close_"+file])
    
    #drop last days entries, for which no return could be computed
    return df.drop(columns=["Close_" + file for file in files]).iloc[:-2]
````

## 4. Data exploration
We cannot use the data from the previous section directly to build an empirical log-return distribution for portfolio optimization task, since the observations do not appear to be independently and identically distributed (i.i.d.), as we will see further below. While it is generally up to the user to decide whether data is i.i.d. or not, there are some obvious situations under which this assumption is difficult to justify. We consider the following criteria and statistical tests to make an informed judgement:

1. The data is stationary. The plots in the previous section show that there is no obvious trend over time in the log returns. However, there may still be a unit root present in the data. We will inspect this by conducting an augmented Dickey-Fuller test on the univariate time series of each asset.

2. The data has no autocorrelation. We will check this by inspecting the (univariate) autocorrelation and partial autocorrelation functions of each asset. Should we identify substantial (partial) autocorrelation, the largest substantial lag might be indicative of the optimal number of parameters in a vector autoregressive moving average (VARMA) model, whose residuals would not show this autocorrelation.

3. The data is homoscedastic, i.e. the variance does not change over time. We will inspect the univariate autocorrelation plots of the squares of each assets mean-subtracted return. Substantial autocorrelation indicates the need to model heteroskedasticity, e.g. by means of a DCC-GARCH model.

We will assume that the data is i.i.d. if all these criteria are met. Consider the following functions that run these three diagnostic tests:

````python
def test_unitRoot(series, signif = 0.05, name = '', verbose = False):
    """Perform augmented Dickey-Fuller test to check for stationarity of given series and print report"""
    r = sm.tsa.stattools.adfuller(series, autolag='AIC')
    output = {'test_statistic':round(r[0], 4), 'pvalue':round(r[1], 4), 'n_lags':round(r[2], 4), 'n_obs':r[3]}
    p_value = output['pvalue'] 

    def adjust(val, length = 6): return str(val).ljust(length)

    # Print Summary
    print(f'    Augmented Dickey-Fuller Test on "{name}"', "\n   ", '-'*47)
    print(f' Null hypothesis: Data has unit root. Non-stationary.')
    print(f' Significance level    = {signif}')
    print(f' Test statistic        = {output["test_statistic"]}')
    print(f' No. lags chosen       = {output["n_lags"]}')

    for key,val in r[4].items():
        print(f' Critical value {adjust(key)} = {round(val, 3)}')

    if p_value <= signif:
        print(f" => P-value = {p_value}. Rejecting null hypothesis.")
        print(f" => Series is stationary.\n")
    else:
        print(f" => P-value = {p_value}. Weak evidence to reject the null hypothesis.")
        print(f" => Series is non-stationary.\n")
        
def test_autoCorr(series, name="", n_std = 1.96, lags=20, figsize=(12,3)):
    """Plots autocorrelation and partial autocorrelation functions next to each other."""
    
    plt.figure(figsize=figsize)
    
    #Autocorrelation function
    lag_acf = sm.tsa.stattools.acf(series.to_numpy(), nlags=lags, fft=True)
    
    #Left subplot   
    plt.subplot(1, 2, 1)
    plt.axhline(y=0,linestyle='--',color='green')
    plt.axhline(y=-1.96/np.sqrt(len(series)),linestyle='-',color='green')
    plt.axhline(y=1.96/np.sqrt(len(series)),linestyle='-',color='green')
    plt.plot(lag_acf, marker="+", markersize=8, color='blue')
    plt.title(f'Autocorrelation function for {name}')
    plt.xlabel('number of lags')
    plt.ylabel('correlation')
    axes = plt.gca()
    axes.set_ylim(1.1*np.minimum(np.min(lag_acf),-1.96/np.sqrt(len(series))),1.1*np.max(lag_acf[1:]));
    plt.tight_layout()
    
    #Partial autocorrelation function
    lag_pacf = sm.tsa.stattools.pacf(series.to_numpy(), nlags=lags)
    
    #Right subplot
    plt.subplot(1, 2, 2)
    plt.axhline(y=0,linestyle='--',color='green')
    plt.axhline(y=-1.96/np.sqrt(len(series)),linestyle='-',color='green')
    plt.axhline(y=1.96/np.sqrt(len(series)),linestyle='-',color='green')
    plt.plot(lag_pacf, marker="+", markersize=8, color='blue')
    plt.title(f'Partial autocorrelation function for {name}')
    plt.xlabel('number of lags')
    plt.ylabel('correlation')
    axes = plt.gca()
    axes.set_ylim(1.1*np.minimum(np.min(lag_pacf),-1.96/np.sqrt(len(series))),1.1*np.max(lag_pacf[1:]));
    plt.tight_layout()
````

## 5. Model selection and estimation

The preceding data exploration section suggests that there should an adequate model of the time series should contain AR and MA terms. Furthermore, it should deal with the heteroskedasticity. To address these two points, we will first train a VARMA model on the data and then fit a Dynamic Conditional Correlation Multivariate Generalized Autoregressive Conditional Heteroskedasticity (DCC-MGARCH(1,1)) model on the residuals of the VARMA model. We will consider multiple parameters, in particular the number of AR- and MA-lags. Since the parameter space is rather compact for this selected model, we will find the optimal model by means of a grid search over the parameters, evaulating the models on a test set in terms of their root-mean squared error (RMSE). For any given set of parameters, a model can be fitted using the following function:

````python
class mgarch:
    
    def __init__(self, dist = 'norm'):
        if dist == 'norm' or dist == 't':
            self.dist = dist
        else: 
            print("Takes pdf name as param: 'norm' or 't'.")
            
    def garch_fit(self, returns):
        res = sp_min( self.garch_loglike, (0.01, 0.01, 0.94), args = returns,
              bounds = ((1e-6, 1), (1e-6, 1), (1e-6, 1)))
        return res.x

    def garch_loglike(self, params, returns):
        T = len(returns)
        var_t = self.garch_var(params, returns)
        LogL = np.sum(-np.log(2*np.pi*var_t)) - np.sum( (returns.A1**2)/(2*var_t))
        return -LogL

    def garch_var(self, params, returns):
        T = len(returns)
        omega = params[0]
        alpha = params[1]
        beta = params[2]
        var_t = np.zeros(T)     
        for i in range(T):
            if i==0:
                var_t[i] = returns[i]**2
            else: 
                var_t[i] = omega + alpha*(returns[i-1]**2) + beta*var_t[i-1]
        return var_t        
        
    def mgarch_loglike(self, params, D_t):
        # No of assets
        a = params[0]
        b = params[1]
        Q_bar = np.cov(self.rt.reshape(self.N, self.T))

        Q_t = np.zeros((self.T,self.N,self.N))
        R_t = np.zeros((self.T,self.N,self.N))
        H_t = np.zeros((self.T,self.N,self.N))
        
        Q_t[0] = np.matmul(self.rt[0].T/2, self.rt[0]/2)

        loglike = 0
        for i in range(1,self.T):
            dts = np.diag(D_t[i])
            dtinv = np.linalg.inv(dts)
            et = dtinv*self.rt[i].T
            Q_t[i] = (1-a-b)*Q_bar + a*(et*et.T) + b*Q_t[i-1]
            qts = np.linalg.inv(np.sqrt(np.diag(np.diag(Q_t[i]))))

            R_t[i] = np.matmul(qts, np.matmul(Q_t[i], qts))


            H_t[i] = np.matmul(dts, np.matmul(R_t[i], dts))   

            loglike = loglike + self.N*np.log(2*np.pi) + 2*np.log(D_t[i].sum()) + np.log(np.linalg.det(R_t[i])) + \
                      np.matmul(self.rt[i], (np.matmul( np.linalg.inv(H_t[i]), self.rt[i].T)))

        return float(loglike)

    
    def mgarch_logliket(self, params, D_t):
        # No of assets
        a = params[0]
        b = params[1]
        dof = params[2]
        Q_bar = np.cov(self.rt.reshape(self.N, self.T))

        Q_t = np.zeros((self.T,self.N,self.N))
        R_t = np.zeros((self.T,self.N,self.N))
        H_t = np.zeros((self.T,self.N,self.N))
        
        Q_t[0] = np.matmul(self.rt[0].T/2, self.rt[0]/2)

        loglike = 0
        for i in range(1,self.T):
            dts = np.diag(D_t[i])
            dtinv = np.linalg.inv(dts)
            et = dtinv*self.rt[i].T
            Q_t[i] = (1-a-b)*Q_bar + a*(et*et.T) + b*Q_t[i-1]
            qts = np.linalg.inv(np.sqrt(np.diag(np.diag(Q_t[i]))))

            R_t[i] = np.matmul(qts, np.matmul(Q_t[i], qts))


            H_t[i] = np.matmul(dts, np.matmul(R_t[i], dts))   

            loglike = loglike + np.log( gamma((self.N+dof)/2.)) - np.log(gamma(dof/2)) \
                      -(self.N/2.)*np.log(np.pi*(dof - 2)) - np.log(np.linalg.det(H_t[i])) \
- ((dof+ self.N)*( ((np.matmul(self.rt[i], (np.matmul( np.linalg.inv(H_t[i]), self.rt[i].T))))/(dof - 2.)) + 1)/2.)

        return -loglike
    
    
    def predict(self, new_rt, ndays = 1, first=True, justH=True):
        if 'a' in dir(self):
            self.rt = np.matrix(new_rt)
            self.T, self.N = self.rt.shape
            
            loglike = 0
            if first == True:
                Q_bar = np.cov(self.rt.reshape(self.N, self.T))

                Q_t = np.zeros((self.T,self.N,self.N))
                R_t = np.zeros((self.T,self.N,self.N)) #matrix of conditional correlations
                H_t = np.zeros((self.T,self.N,self.N)) #conditional covariance matrix
                Q_t[0] = np.matmul(self.rt[0].T/2, self.rt[0]/2)
                
                D_t = np.zeros((self.T, self.N))
                var = np.zeros((self.T, self.N))
                for i in range(self.N):
                    params = self.params[i]
                    var[:,i] = self.garch_var(params, self.rt[:,i])
                    D_t[:,i] = np.sqrt(var[:,i])
                self.D_t = D_t
                self.var = var
                
                for i in range(1,self.T):
                    dts = np.diag(self.D_t[i])  #diagonal matrix of conditional variances
                    dtinv = np.linalg.inv(dts)  #its inverse
                    et = dtinv*self.rt[i].T     
                    Q_t[i] = (1-self.a-self.b)*Q_bar + self.a*(et*et.T) + self.b*Q_t[i-1]
                    qts = np.linalg.inv(np.sqrt(np.diag(np.diag(Q_t[i]))))
                    R_t[i] = np.matmul(qts, np.matmul(Q_t[i], qts))
                    H_t[i] = np.matmul(dts, np.matmul(R_t[i], dts))
                self.Q_t=Q_t
                self.R_t=R_t
                self.H_t=H_t
                self.Q_t_last1=Q_t[-1]

            else:
                D_t_last = np.zeros((self.N,))
                for i in range(self.N):
                    params = self.params[i]
                    D_t_last[i] = np.sqrt(params[0] + params[1]*(self.rt[-2,i]**2) + params[2]*var[-2,i])
                
                #create storage
                Q_bar = np.cov(self.rt.reshape(self.N, self.T))
                
                #add last entry
                i=self.T-1 #last index is one less than number of time steps
                dts = np.diag(self.D_t_last)  #diagonal matrix of conditional variances
                dtinv = np.linalg.inv(dts)  #its inverse
                et = dtinv*self.rt[i].T     
                Q_t_last = (1-self.a-self.b)*Q_bar + self.a*(et*et.T) + self.b*self.Q_t_last1
                qts = np.linalg.inv(np.sqrt(np.diag(np.diag(Q_t_last))))
                R_t_last = np.matmul(qts, np.matmul(Q_t_last, qts))
                H_t_last = np.matmul(dts, np.matmul(R_t_last, dts))
                self.Q_t_last1=Q_t_last
                
            #additional return options
            if justH == True:
                return H_t[-1, :, :]
            if self.dist == 'norm':
                return {'dist': self.dist, 'cov': H_t[-1]*np.sqrt(ndays)}
            elif self.dist == 't':
                return {'dist': self.dist, 'dof': self.dof, 'cov': H_t[-1]*np.sqrt(ndays)}
            
        else:
            print('Model not fit')
            
            
    def fit(self, returns):
        self.rt = np.matrix(returns)
        
        self.T = self.rt.shape[0]
        self.N = self.rt.shape[1]
        
        if self.N == 1 or self.T == 1:
            return 'Required: 2d-array with columns > 2' 
        self.mean = self.rt.mean(axis = 0)
        self.rt = self.rt - self.mean
        
        D_t = np.zeros((self.T, self.N))
        params=[]
        for i in range(self.N):
            params.append(self.garch_fit(self.rt[:,i]))
            D_t[:,i] = np.sqrt(self.garch_var(params[-1], self.rt[:,i]))
        self.D_t = D_t
        self.params = params
        if self.dist == 'norm':
            res = sp_min(self.mgarch_loglike, (0.01, 0.94), 
                           args = D_t, bounds = ((1e-6, 1), (1e-6, 1)))#,options = {'maxiter':10000000, 'disp':True})
            self.a = res.x[0]
            self.b = res.x[1]
            
            return {'mu': self.mean, 'alpha': self.a, 'beta': self.b} 
        elif self.dist == 't':
            res = sp_min(self.mgarch_logliket, (0.01, 0.94, 3), args = D_t,
            bounds = ((1e-6, 1), (1e-6, 1), (3, None)), 
            #options = {'maxiter':10000000, 'disp':True},
            )
            self.a = res.x[0]
            self.b = res.x[1]
            self.dof = res.x[2]
            return {'mu': self.mean, 'alpha': self.a, 'beta': self.b, 'dof': self.dof}
````

````python
def model_fit(X_train, p, q):
    """Fits a VARMA model of order (p,q) on the training data in X_train"""
    model_varma = sm.tsa.VARMAX(X_train, order=(p, q))
    result_varma = model_varma.fit(maxiter=100, disp=False)    
    return result_varma

def model_getResid(model_dcc, resid_in):
    """Compute residuals of the DCC-MGARCH model"""
    model_dcc.predict(new_rt = resid_in) #called to obtain model_dcc.H_t
    
    #Compute variance of output residuals
    H_t_half_inv = np.zeros((model_dcc.T, model_dcc.N, model_dcc.N)) #allocate storage
    for tt in range(1, model_dcc.T):
        H_t_half_inv[tt,:,:] = inv(cholesky(model_dcc.H_t[tt,:,:]))
    
    #Compute and return output residuals z
    z = np.zeros((model_dcc.T,model_dcc.N)) #allocate storage
    for tt in range(model_dcc.T):
        z[tt,:] = H_t_half_inv[tt,:,:] @ (resid_in[tt,:]-np.array(model_dcc.mean).reshape(model_dcc.N,))
    
    return z


def model_test(result_varma, X_test):    
    """Returns training and testing predictions for the VARMA model"""
    
    #Training predictions
    pred_horizon = len(X_test)
    Y_train = result_varma.predict()
    
    #Testing predictions
    result_varma = result_varma.extend(X_test)
    Y_test = result_varma.predict()
    
    return Y_train, Y_test

def sample_pred(ret_hist, rand_z, pred_horizon, res_varma_in, dcc_in, varma_range=1):
    """Return a sample path of predictions for the combined VARMA and DCC-MGARCH model."""
    
    dcc = copy.copy(dcc_in)
    T = copy.copy(dcc.T) #length of original time series
    res_varma = copy.copy(res_varma_in)
    dcc_mean = np.array(dcc.mean).reshape(dcc.N,)
    resid_varma = np.r_[res_varma.resid, np.zeros((pred_horizon,dcc.N))]  #residuals of the VARMA model
    pred_varma = np.r_[res_varma.predict(), np.zeros((pred_horizon,dcc.N))] #predicted returns of the VARMA model
    pred_ret = np.r_[ret_hist,np.zeros((pred_horizon,dcc.N))] #predicted returns of the combined VARMA and DCC-MGARCH model
    
    #Go through prediction hoerizon
    for ttp in range(pred_horizon):
        #update prediction with extended model
        if ttp==0:
            pred_varma[T+ttp,:] = res_varma.predict(start = T + 1 + ttp, end = T + 1 + ttp)
        else:
            pred_varma[T+ttp,:] = res_varma.predict(start = varma_range + 2, end = varma_range+2) #+1 for index +1 for 0th order term

        #predict conditional covariance matrix
        H_t1 = dcc.predict(justH = True, first = ttp == 0, new_rt = resid_varma[:(T+ttp),:])

        #predict new data point
        resid_varma[T+ttp,:] = dcc_mean + cholesky(H_t1) @ rand_z[ttp, :]
        pred_ret[T+ttp,:] = pred_varma[T+ttp, :] + resid_varma[T+ttp, :]
        
        #extend VARMA model only to the new and old data points that are relevant for prediction in the next loop iteration
        res_varma = res_varma.extend(pred_ret[T+ttp-varma_range:T+ttp+1,:]) #add new return value to VARMA results
        
    return pred_ret[-pred_horizon:,:]
````

## 8. Mean-VaR portfolio optimization via MILP

We now present a portfolio optimization technique that seeks to maximize the expected return while accounting for a finite number of constraints related to limits of the value at risk (VaR) at one or several significance levels. We will show that this formulation makes it possible to view portfolio optimization as maximizing the mean of the distribution of predicted returns, that is linearly parametrized by the portfolio weights, subject to keeping the left/negative tail of the distribution below a user-defined distribution, which we will refer to as the _risk profile_. We will then bring this problem into the form of an MILP problem.

Let $\rho \in \mathbb{R}^d, d \in \mathbb{N}$ denote a random variable describing the returns of $d$ assets. Let $D:=\{1,2,\dots,d\}$ and let $w\in W:=[0,1]^d$, such that $\sum_{i\in D} w_i = 1$, denote the weights of the assets in the portfolio. Then the (random) portfolio return is $\mathbb{E}[\rho]^\top w$, where $\mathbb{E}[\cdot]$ denotes the (element-wise) expectation operator associated with the distribution of each of the $d$ assets.

Since the exact expectation is unavailable in practice, we replace $\mathbb{E}[\rho]$ in the expression above by $\bar{\rho}\in \mathbb{R}^d$, which denotes the empirical mean return of $\rho$ from $n\in \mathbb{N}$ observations $\rho^{(n)}$, i.e. $\bar{\rho}_i = \frac{1}{n}\sum_n \rho_i^{(n)}$. Hence, our maximisation objective becomes 
\begin{equation}
\underset{w\in W}{\max}\bar{\rho}^\top w,
\end{equation}
which is linear in $w$. Let us now focus on the additional constraints that we seek to enforce. To this end, define two vector-valued parameters, $A\in [0,1]^k$ and $V\in\mathbb{R}^k$, where $k\in \mathbb{N}$. These are significance levels and corresponding values at risk. For example $A:=[0.01,0.05]$ and $V:=[-0.2,-0.1]$ would denote two VaR constraints, i.e. the 1% value at risk should be no smaller than -20% and the 5% value at risk should be no smaller than -10%. The constraints are then formulated, using an additional auxiliary variable $z^{(a)}\in [0,1]^n$ for all $a \in K:=\{1,2,\dots,k\}$, as

\begin{align}
\sum_{i\in N} z^{(a)}_i =& n A_a \text{, for all } a \in K,\\
\bar{\rho}_i^\top w \geq& V_a - M(1-z^{(a)}_i) \text{, for all } (a,i) \in K\times N,
\end{align}

where $M\in \mathbb{R}_++$ is sufficiently large, such that the inequality is always satisfied when $z^{(a)}_i=1$, for all choices of $w\in W$. Hence, this reformulation is also known as the big $M$ method. The first equality constraint ensures that for any VaR condition $a\in K$, the number of $z^{(a)}_i=1$ equal the fraction of samples $n$, i.e. $z^{(a)}$ is a selector variable that ensures that sufficiently many constraints need to be satisfied, i.e. if $z^{(a)}_i =1$, then the weights $w$ must be chosen such that $\bar{\rho}_i^\top w \geq V_a$. This ensures that the VaR constraint is satisfied. Combining the objective and the constraints, we are now ready to implement this problem in Python.

We use Pyhon and the $\texttt{mip}$ package (https://www.python-mip.com/) for mixed-integer linear programming and formulate the optimization routine by means of the following function:

````python
def portfolioMILP(X, alpha=[0.05], VaR=[0], M = None, min_weight=None, max_weight=None, max_s = 10, abs_gap=1e-3):
    """
    Optimizes portfolio weights in a Mixed-Integer Linear Programming (MILP) setting, based on data X with n rows 
    (observations) and d columns (assets), given a list (even if only a single value is desired!) of values at risk (VaR) 
    at significance levels alpha, respectively. The variable "M" is used in a big M method trick to linearise the 
    constraints. "max_weights" defines the maximum weight of an asset, "max_s" defines the maximum computation time, 
    "abs_gap" defines the absolute gap between upper bound and lower bound on the solution, upon which the algorithm shall 
    terminate.
    """
    
    n,d = X.shape
    X_mean = X.mean(axis=0)
    print(f"Optimizing over {d} assets using {n} samples")
    
    n_alpha = len(alpha)
    V=[[] for i in range(n_alpha)]
    #Compute "big M" based on data if not provided:
    if M is None:
        M = np.max(X)-np.min(X) #ensures that 
    m = Model(solver_name=CBC)
    m.max_mip_gap_abs = abs_gap
    
    w = [ m.add_var(var_type=CONTINUOUS) for dd in range(d) ] #decision variables (asset weights)
    m.objective = mip.minimize(-xsum(w[dd]*X_mean[dd] for dd in range(d))) #objective function
    m += xsum(w[dd] for dd in range(d)) == 1 #asset weights sum to 1
    
    #VaR constraints
    for aa in range(n_alpha): #for all VaR constraints
        n_activeCon = np.ceil(n*(1-alpha[aa])) #number of constraints that need to be active at solution, i.e. satisfied
        
        v = [] #indices of violation samples (samples for which a feasible w exists that would violate the VaR constraint)
        for i in range(n):
            if any(X[i,:] < VaR[aa]): #else, no weights w lead to a return of sample i to be less than the VaR
                v.append(i)
        n_v = len(v) 
        V[aa] = v #useful for visualising the solution
        
        #selector variable for active constraints, i.e. those that are satisfied
        z = [m.add_var(name = f"z_{aa}_{vv}",var_type = BINARY) for vv in range(n_v)] 
        
        #n-n_v constraints always satisfied, hence consider n_activeCon - (n - n_v)
        m += xsum(m.var_by_name(f'z_{aa}_{vv}') for vv in range(n_v)) == max(0, n_activeCon - (n - n_v))
        for vv in range(n_v):
            #linearized VaR constraint using big M notation
            m += xsum(w[dd]*X[v[vv],dd] for dd in range(d)) >= (VaR[aa] - (1-m.var_by_name(f'z_{aa}_{vv}'))*M) 
    
    #weight constraints
    #maximum weights
    if max_weight is None:
        pass #don't add any constraints if none are specified
    elif type(max_weight) != list and max_weight != 1: 
        for dd in range(d):
            m += w[dd] <= max_weight
    else:
        for dd in range(d):
            m += w[dd] <= max_weight[dd]
    
    #minimum weights
    if min_weight is None:
        pass #don't add any constraints if none are specified
    elif type(min_weight) != list and max_weight != 1: 
        for dd in range(d):
            m += w[dd] >= min_weight
    else:
        for dd in range(d):
            m += w[dd] >= min_weight[dd]

    #obtain model summary
    m.write('model.lp')
    m.read('model.lp')
    print('The model has {} vars, {} constraints and {} nzs'.format(m.num_cols, m.num_rows, m.num_nz))

    # optimizing
    status = m.optimize(max_seconds = max_s)
    if status != OptimizationStatus.OPTIMAL:
        print(np.round(alpha,3), np.round(VaR,3), status) #if solution not optimal, print details
    return m,V
````

Let's run a simple example using the i.i.d. data obtained earlier:

````python
alpha=[0.01, 0.05, 0.20]
VaR=[-0.2, -0.1, 0]
m,V = portfolioMILP(rho, alpha, VaR, min_weight = 0.1, max_s = 1800) #run MILP for at most max_s seconds

if m.status == OptimizationStatus.OPTIMAL or m.status == OptimizationStatus.FEASIBLE:
    wStar = [m.vars[i].x for i in range(np.shape(rho)[1])] #optimal weights
    print(m.status)
    print("The optimal asset allocation is: " \
          + str([files[i] + ": " + str(np.round(wStar[i]*100,2)) + " %" for i in range(len(files))]))
    print("Portfolio mean return: "+str(np.round(sum(wStar*rho.mean(axis=0))*100,2))+" %")
    print("Portfolio std: "+str(np.round(np.std([sum(wStar*rho[i,:]) for i in range(rho.shape[0])])*100,2))+" %")
````

Often, the solution will only have status`FEASIBLE`instead of`OPTIMAL`for small maximum computation times`max_s`and when the problem size is large. We can now visualize the result to see how the MILP algorithm has segmented the distribution according to the VaR constraints:

````python
def plotDistribution(X, m, V, alpha, i0=0, i1=1, names=["rho_1", "rho2"], xlim = None, ylim = None, \
                     cmap_str = 'gist_rainbow', markersize=2):
    """Scatter plot of the empirical distributions of X along axes i0 and i1. Color-coded according to which samples 
    violate the VaR conditions"""
    
    #Extract data
    x1=X[:,i0]
    x2=X[:,i1]
    
    n_x=len(x1)
    n_alpha=len(alpha)
    
    #Compute find indices of samples that violate VaR conditions (dropped) or do not violate them (kept) for color-coding
    dropped=[[] for i in range(n_alpha)]
    kept=list(range(n_x))
    for aa in range(n_alpha):
        for vv in range(len(V[aa])):
            if m.var_by_name(f'z_{aa}_{vv}').x == 0:
                dropped[aa].append(V[aa][vv])

    #Plot figure
    cmap=plt.cm.get_cmap(cmap_str,n_alpha)
    
    plt.scatter(x1[kept],x2[kept],s=markersize,c='grey')
    for aa in reversed(range(n_alpha)):
        plt.scatter(x1[dropped[aa]],x2[dropped[aa]],s=markersize,color=cmap(aa/(n_alpha)))
    
    #Figure formatting
    plt.grid(True)
    plt.xlabel(names[0])
    plt.ylabel(names[1])
    #plt.title('Annual returns')
    plt.gca().set_aspect('equal', adjustable='box')
    if xlim is not None and ylim is not None:
        axes = plt.gca()
        axes.set_xlim(xlim)
        axes.set_ylim(ylim)
    return
````

Finally, we can plot the predicted portfolio distribution:

````python
def getEmpiricalDistribution(data, xlim=None):
    """Compute and plot empirical cumulative distributions of data"""
    
    #Compute data dimensions
    n=data.shape[0]
    try:
        data.shape[1]
    except IndexError:
        data = data.reshape(n,1)
    d=data.shape[1]
    
    #Compute auxiliary data frame
    df2=pd.DataFrame()
    for i in range(d):
        df2[f"value_{i}"]=data[:,i]
        df2[f"frequency_{i}"] = 1 #absolute frequency
        df2[f'pdf_{i}'] = df2[f'frequency_{i}'] / sum(df2[f'frequency_{i}']) #probability density function
        df2 = df2.sort_values(f"value_{i}") 
        df2[f'cdf_{i}'] = df2[f'pdf_{i}'].cumsum() #cumulative distribution function

    #Plot results
    fig, ax = plt.subplots(figsize=(8,8))
    for i in range(d):
        df2 = df2.sort_values(f"value_{i}")
        if xlim is not None:
            df2.plot(ax = ax, x = f'value_{i}', y = f'cdf_{i}', xlim=xlim, drawstyle="steps-post", grid = True, legend=False, linewidth=2, color='black')
        else:
            df2.plot(ax = ax, x = f'value_{i}', y = f'cdf_{i}', grid = True,  drawstyle="steps-post", legend=False, linewidth=2, color='black')
````

## 9. Conclusion

In this Jupyter notebook, we explored an alternative asset allocation approach to modern portfolio. We built time series models to obtain bootstrapped predictions of the returns of several assets. This allowed us to obtain an empirical distribution of predicted returns, which we used to formulate a mixed-integer linear programming problem to optimally allocate assets in the portfolio.

Some open questions remain:
1. How scalable is the forecasting approach? Especially forecasting multiple sample paths in the current implementation is computationally expensive. Improvements of bootstrapped prediction methods in the statsmodels and mgarch packages could speed up this process. For a large-scale implementation of this methodology, it also seems reasonable to parallelize the forcecasting process. This is possible, since the forecasts are independent of each other.
2. How scalable is the MILP procedure? Even though the number of non-zeros in the MILP grows at a modest linear rate with the number of VaR constraints, the computation becomes costly fast if one also wants to optimize over more than a handful of assets. Alternative solver and parallel computation might provide sufficent improvements for large-scale problems.
3. How statistically robust is the overall approach? In the above approach, we are first sampling residuals of our time series model and then optimizing over the resulting distribution of predicted returns. At both of these stages, we implicitly assume that the available distributions, $z$ and $\rho$ correspond to the "true" distribution. However, these are only (i.i.d.) samples from these "true" distributions. To obtain solid, probabilistic results, one could bound the gap between the sampled distributions and the "true" distributions using, for example, the Dvoretzky-Kiefer-Wolfowitz bound.
